# Q1. Explain GET and POST methods.

GET and POST are two common HTTP methods used to submit data from a client to a server.

The GET method is used to retrieve data from a server. When a client sends a GET request to a server, the server retrieves the specified data and sends it back to the client. The data can be specified in the URL, as query parameters.

The POST method, on the other hand, is used to send data to a server to create or update a resource. When a client sends a POST request to a server, the server receives the data in the request body and processes it.

# Q2. Why is request used in Flask?

The request object is a very important object in Flask, and it is used to access data submitted in an HTTP request.

When a client sends an HTTP request to a Flask application, the Flask framework automatically creates a request object that contains all the information about the request, such as the request method (GET, POST, etc.), headers, form data, query parameters, and more.

# Q3. Why is redirect() used in Flask?

The redirect() function in Flask is used to redirect a client to a different URL.

In web applications, it's common to redirect users to different pages after they've completed an action, such as submitting a form or logging in. The redirect() function provides a simple way to do this in Flask.

# Q4. What are templates in Flask? Why is the render_template() function used?

Templates in Flask are files that contain the structure and layout of a web page or email message. In Flask, templates are typically written in HTML with placeholders for dynamic content that is generated by the application.

The render_template() function in Flask is used to generate a complete HTML page using a template file and some dynamic data provided by the application. The function takes the name of a template file as its first argument and any additional data that needs to be passed to the template as keyword arguments.
